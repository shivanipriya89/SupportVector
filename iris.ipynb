```python
#5-** Split your data into a training set and a testing set.**
```

```python
df # Original data frame
```

```python
df["species"]=pd.get_dummies(df["species"])
```

```python
df["species"]
```

```python
df # Refined data frame
```

```python
x=df[["sepal_length","sepal_width","petal_length","petal_width"]]
```

```python
y=df["species"]
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#6-**Call the SVC() model from sklearn and fit the model to the training data.**
```

```python
import sklearn
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
#7-Now get predictions from the model and create a confusion matrix and a classification report.**
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
import sklearn
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
